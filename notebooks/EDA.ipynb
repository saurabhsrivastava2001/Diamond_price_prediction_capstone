```python
import pandas as pd

# Load your second dataset
df2 = pd.read_csv(r"D:\Desktop\DiamondPricePrediction-main\notebooks\data\train2.csv")

# Define a dictionary to map the second dataset's columns to the first dataset's columns
column_mapping = {
    "Carat(Weight of Daimond)": "carat",
    "Cut(Quality)": "cut",
    "Color": "color",
    "Clarity": "clarity",
    "Depth": "depth",
    "Table": "table",
    "Price(in US dollars)": "price",
    "X(length)": "x",
    "Y(width)": "y",
    "Z(Depth)": "z"
}

# Rename the columns
df2 = df2.rename(columns=column_mapping)

# Now, df2 will have the same columns as your first dataset
print(df2.head())  # Display the first few rows to check the new column names

# Optionally, save the modified dataset to a new CSV file
df2.to_csv("renamed_second_dataset.csv", index=False)
```

## encoding categorical columns --

```python
data["cut"].unique()
```

```python
data["clarity"].unique()
```

```python
data["color"].unique()
```

```python
#ORDINAL ENCODING

cut_map= {"Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5}

clarity_map= {"I1":1,"SI2":2,"SI1":3,"VS2":4,"VS1":5,"VVS2":6,"VVS1":7,"IF":8}

color_map= {"D":1,"E":2,"F":3,"G":4,"H":5,"I":6,"J":7}
```

```python
data["cut"]=data["cut"].map(cut_map)
```

```python
data["clarity"]=data["clarity"].map(clarity_map)
```

```python
data["color"]=data["color"].map(color_map)
```
